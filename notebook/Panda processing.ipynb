```python
import sys
import os 
import pandas as pd
import numpy as np
import torch
import glob
import matplotlib.pyplot as plt
from  sklearn.preprocessing import minmax_scale
```

```python
path = r'/home/lburget/Documents/EPFL/Master/PDS2/Lucas/Gesture/Data2/'
all_files = glob.glob(os.path.join(path, "*.csv")) 
```

Concatenate all files in the different folders into one single panda data frame. 

```python
all_file_list = []
for r, dirs, files in os.walk(path):
    for d in dirs:
        path2 = path + d 
        all_files = glob.glob(os.path.join(path2, "*.csv"))
        for f in all_files:
            prov = pd.read_csv(f)
            prov['gesture_nbr'] = f.split(path2+'/')[1][:-4]
            prov['label'] = d
            all_file_list.append(prov)
df_all = pd.concat(all_file_list, axis=0)
```

```python
df_all.head()
```

## Preprocessing

```python
df_all['rospy_time'] = pd.to_datetime(df_all['rospy_time'], unit='ns')
df_all.head()
```

```python
df_all.set_index(['label', 'gesture_nbr', 'rospy_time'], inplace=True)
df_all.head()
```

Sort all the gestures per increasing time, to ensure that the order is meaningful. 

```python
df2 =  df_all.groupby(['label', 'gesture_nbr'], group_keys=False).apply(lambda x: x.sort_values('rospy_time'))
df2.head()
```

Downsample every 20ms to ensure equally spaced values over the dataset.

```python
level_values = df2.index.get_level_values
df3 = (df2.groupby([level_values(i) for i in [0,1]]
                      +[pd.Grouper(freq='20L', level=-1)]).mean().fillna(method = 'bfill').rolling(3,1,center = True).mean())
df3.head()
```

Create a columns "energy" computed with the norm of the acceleration. It is used to know when the gestures starts. 

```python
df3['energy'] = 0
column_meaning = ['accel_right_x', 'accel_right_y', u'accel_right_z', u'accel_left_x',
       u'accel_left_y', u'accel_left_z']

df3['energy'] =  df3.groupby(['label', 'gesture_nbr'], group_keys=False).apply(lambda x: x.loc[:,column_meaning].pow(2).sum(axis = 1).apply(np.sqrt)/6)
```

Find index at which energy is over a treshold and extract 1.5s from there if possible.

```python
l = []
for truc, much in df3.groupby(['label', 'gesture_nbr'], group_keys=False):
    new = much.loc[:,column_meaning].pow(2).sum(axis = 1).apply(np.sqrt)/6
    min_idx  = list(new[new>0.04].index.min()) 
    end_idx  = list(new[new>0.04].index.min()) 
    end_idx[2]= end_idx[2] + pd.Timedelta('1.5s')
    l.append(much.loc[tuple(min_idx):tuple(end_idx)])
    print(much.loc[tuple(min_idx):tuple(end_idx)].shape)
df4 = pd.concat(l, axis=0)
```

```python
df4.head()
```

Take the difference between two time steps. This will allow the model to learn the relative pattern and note the absolute one.

```python
df5 = df4.groupby(['label', 'gesture_nbr'], group_keys=False).diff().fillna(0.0)
for truc, much in df5.loc['aggregate'].groupby(['gesture_nbr'], group_keys=False):
    ax = much['gyro_left_z'].rolling(5,1).mean().reset_index(drop=True).plot(y='gyro_left_z')
    # print(much['accel_right_y'].rolling(3,1).mean().)
ax.set_ylabel('Acceleration change')
ax.set_xlabel('Index (1.5s long)')
ax.set_title('Gyroscope z left')
```

Down-sample again to 40ms

```python
level_values = df5.index.get_level_values
df6 = (df5.groupby([level_values(i) for i in [0,1]]
                      +[pd.Grouper(freq='40L', level=-1)]).mean().fillna(method = 'bfill'))
df6.head()
```

```python
for truc, much in df6.loc['aggregate'].groupby(['gesture_nbr'], group_keys=False):
    much['accel_left_x'].rolling(5,1, center = True).mean().reset_index(drop=True).plot(y='accel_right_x')
```

Split in test and train set.

```python
from sklearn.model_selection import train_test_split

test_l = []
train_l = []
for label, df in df6.groupby(['label']) :
    unique_nbr = df.loc[label].reset_index()['gesture_nbr'].unique()
    ratio = int(len(unique_nbr)*0.2)
    test_choice = np.random.choice(unique_nbr, ratio, replace=False)
    train_choice = set(unique_nbr)-set(test_choice)
    test_l.append(df.loc[label, test_choice, :])
    train_l.append(df.loc[label, train_choice, :])
df_test = pd.concat(test_l, axis=0)
df_train = pd.concat(train_l, axis =0)
df_train.head()
```

Scale to [-1,1] and save values for later use.

```python
from sklearn import preprocessing

columns_to_norm = column_meaning = ['accel_right_x', 'accel_right_y', u'accel_right_z', u'accel_left_x',
       u'accel_left_y', u'accel_left_z', u'gyro_right_x', u'gyro_right_y',
       u'gyro_right_z', u'gyro_left_x', u'gyro_left_y', u'gyro_left_z']

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))
param_norm = dict()
for name in columns_to_norm:
    df_train[name] = min_max_scaler.fit_transform(df_train[name].values.astype(float).reshape(-1,1))
    param_norm[name] = [ min_max_scaler.data_min_, min_max_scaler.data_max_]
```

Select the longest sequence of input.

```python
size_max = df_train.drop('energy', axis = 1).groupby(['label', 'gesture_nbr']).apply(lambda x : x.shape).max()
```

Check the number of different gestures (classes).

```python
NBR_CLASS = len(df_train.index.levels[0].values)
```

Fill torch.Tensor with the values of the input sequence for the 20 dimensions. All the tensors have the same shape (20, max_length). The input is padded with 0 if it is too short.

```python
N_FEATURES = 20

segments = []
labels = []

for label, df in df_train.groupby(['label', 'gesture_nbr']):
    target = torch.zeros((size_max[1],size_max[0]))
    max_size = df.drop('energy', axis = 1).values.shape[0]
    #print(np.transpose(df.drop('energy', axis = 1).values).shape)
    #print(torch.tensor(np.transpose(df.drop('energy', axis = 1).values)))
    target[:,:max_size] = torch.tensor(np.transpose(df.drop('energy', axis = 1).values))
    #print(torch_tensor.shape)
    segments.append(target)
    labels.append(label[0])

# Bring the segments into a better shape
target_reshaped = torch.cat(segments)
target_reshaped.shape
```

```python
target_reshaped = target_reshaped.reshape(-1,N_FEATURES, size_max[0])#  .asarray(segments, dtype= np.float32).reshape(-1, time_steps, N_FEATURES)
target_reshaped.shape
```

Transform gesture's name into label. 

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(labels)
le.classes_
# le.inverse_transform([0, 0, 1, 2])
label_encoded = torch.tensor(le.transform(labels))
label_encoded
```

Shuffle the tensor to permute the sample.

```python
idx = torch.randperm(len(label_encoded))
target_reshaped = target_reshaped[idx, :, :]
label_encoded = label_encoded[idx]
```

```python
N_FEATURES = 20

segments = []
labels = []

for label, df in df_test.groupby(['label', 'gesture_nbr']):
    target = torch.zeros((size_max[1],size_max[0]))
    max_size = df.drop('energy', axis = 1).values.shape[0]
    target[:,:max_size] = torch.tensor(np.transpose(df.drop('energy', axis = 1).values))
    #print(torch_tensor.shape)
    segments.append(target)
    labels.append(label[0])

# Bring the segments into a better shape
test_reshaped = torch.cat(segments)
test_reshaped = test_reshaped.reshape(-1,N_FEATURES, size_max[0])#  .asarray(segments, dtype= np.float32).reshape(-1, time_steps, N_FEATURES)
test_reshaped.shape
```

```python
label_encoded_test = torch.tensor(le.transform(labels))
```

## 1D CNN network

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv1d(20, 5, 6) # out  batch_s * 10 * max_size -5 = 34
        self.pool = nn.MaxPool1d(2) # batch_s * 5 * 17 
        # self.conv2 = nn.Conv1d(10, 5, 10) # batch_s* 5 * 24
        self.conv3 = nn.Conv1d(5, 2, 10) # batch_s * 2 * 8
        self.pool2 = nn.AvgPool1d(2) # batch_s * 2 * 4
        
        self.fc1 = nn.Linear(2 * 4, 20)
        self.fc2 = nn.Linear(20, 10)
        self.fc3 = nn.Linear(10, NBR_CLASS)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x))) # 
        # x = F.relu(self.conv2(x))
        x = self.pool2(F.relu(self.conv3(x)))

        x = x.view(-1, 2 * 4)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
import torch.optim as optim

net = Net()

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters())
mini_batch_size = 10
cEL = []
min_current = 1000000000000

lambda_l1  =  0.0001
 
for epoch in range(100):  # loop over the dataset multiple times

    running_loss = 0.0
    for b in range(0, target_reshaped[:60,:,:].size(0), mini_batch_size):
        # get the inputs; data is a list of [inputs, labels]

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        output = net(target_reshaped[:60,:,:].narrow(0, b, mini_batch_size)) # dim,start,length
        loss = criterion(output, label_encoded[:60].narrow(0, b, mini_batch_size))
        loss.backward()
        optimizer.step()
        
        with torch.no_grad():
            for p in net.parameters():
                p.sub_(p.sign() * p.abs().clamp(max = lambda_l1))


        # print statistics
        running_loss += loss.item()
        
        with torch.no_grad():
            out_test = net(test_reshaped)
            if F.cross_entropy(out_test, label_encoded_test)< min_current:
                best_param = net.state_dict()
                min_current = F.cross_entropy(out_test, label_encoded_test)
          
        if b % 10 == 0:    # print every 2000 mini-batches
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, b + 1, running_loss /10 ))
            running_loss = 0.0

print('Finished Training')
```
